# HE $\gamma$-ray long-term light-curve of BL Lacertae

In this notebook, we study the properties of the high-energy (HE; $100\text{MeV}<E<100\text{GeV}$) long-term (2008-2024) light-curve of BL Lacertae with bins of 3 days, with the aim of simulating light-curves having the same flux probability density function (PDF) and same power spectral density (PSD) as the observed one.

The light-curve simulations are performed using the method described in Emmanoulopoulos D. et al. 2013, *Monthly Notices of the Royal Astronomical Society*, Volume 433, Issue 2, Pages 907â€“927 [https://academic.oup.com/mnras/article/433/2/907/1746942]. 

The method was implemented by Sam Connolly in the DELCgen package, available at https://github.com/samconnolly/DELightcurveSimulation. Here we use DELCgen3, which is a Python 3 version of the code, available at https://github.com/cerasole/DELightcurveSimulation.

Reference: **S. Abdollahi et al. 2023, "The Fermi-LAT Lightcurve Repository", The Astrophysical Journal Supplement Series (ApJS), Volume 265, Number 2, 31** [https://www.aanda.org/articles/aa/full_html/2018/12/aa33621-18/aa33621-18.html]

The data is publicly available at [https://fermi.gsfc.nasa.gov/ssc/data/access/lat/LightCurveRepository/source.html?source_name=4FGL_J2202.7+4216]. The data files are not included in this repository, they are to be downloaded from the server.

```python
### Basic imports
import numpy as np
import pandas as pd
import datetime
from astropy.time import Time
import astropy.units as u
import scipy.stats as st
import matplotlib.pyplot as plt

### DELCgen3
import os, sys
sys.path.append("../../DELCgen3/")
from DELCgen3 import Lightcurve, OptBins, Mixture_Dist, BendingPL, PL_no_offset, PL_with_offset, Simulate_TK_Lightcurve, Simulate_DE_Lightcurve

# Style
fontsize = 16
plt.rc("font", family = "Dejavu Sans", weight = "normal", size = fontsize)

mjd_time_ref = datetime.datetime(1858, 11, 17)
def date2MJD (date):
    try:
        return (date - mjd_time_ref).days
    except:
        return np.array([date2MJD(date_) for date_ in date])
```

### Load the data

```python
filename = "../../Data/4FGL_J2202.7+4216_daily_02_11_2024.csv"

def get_FermiLAT_data (filename, TS_cut = 4):
    data = pd.read_csv(filename)
    uplims = data["TS"] < TS_cut
    mjd = data["Julian Date"][~uplims].to_numpy().astype("float32") - 2400000.5
    flux = data["Photon Flux [0.1-100 GeV](photons cm-2 s-1)"][~uplims].to_numpy().astype("float32") * u.Unit("cm-2 s-1")
    flux_err = data["Photon Flux Error(photons cm-2 s-1)"][~uplims].to_numpy().astype("float32") * u.Unit("cm-2 s-1")
    return mjd, flux, flux_err

mjd, flux, flux_err = get_FermiLAT_data(filename)
```

### Definition of the Lightcurve object and plot of the light-curve.

Caveats:
- in the DELCgen code, although the `Lightcurve` object can take as argument the vector of uncertainties of the fluxes, these uncertainties are not used anywhere in the simulation. They are only used when in the `Lightcurve.Plot_Lightcurve()` method.
- in the DELCgen documentation, it is specified that the light-curve has to be evenly binned. The bin width of the light-curve is given by the tbin argument, which is mandatory. The Periodogram and PSD are computed assuming the light-curve is defined on an evenly-spaced time grid. However, optical light-curves are necessarily unevenly binned. This is an intrinsic limitation of the DELCgen code.

```python
lc_gamma = Lightcurve(mjd, flux.to_value("1e-6 cm-2 s-1"), tbin = 3)
fig_lc, ax_lc = lc_gamma.Plot_Lightcurve()
lc_name = "BL Lac HE $\gamma$-ray light-curve in $2008-2025$"
ax_lc.set_title(lc_name)
ax_lc.set_xlabel("Time [MJD]");
ax_lc.set_ylabel("Photon Flux ($0.1$ GeV$<$E$<100$ GeV)\n[$10^{-6}$ cm$^{-2}$ s$^{-1}$]")
ax_lc.grid();

ax = ax_lc
xticks = np.arange(55000, 61000, 1000)
ax.set_xticks(xticks)
date_top_x_axis = np.array([datetime.datetime(year, 1, 1, 0, 0, 0) for year in range(2008, 2026)])
mjd_top_x_axis = date2MJD(date_top_x_axis)
ax.set_xlim(left = mjd_top_x_axis[0], right = mjd_top_x_axis[-1])
date = Time(xticks, format='mjd').datetime
dates = [date_top_x_axis[i].strftime('20%y') for i in range (0, len(date_top_x_axis))]
ax_top = ax.twiny()
ax_top.set_xlim(ax.get_xlim())
ax_top.set_xticks(mjd_top_x_axis)
dates_top_axis = ax_top.set_xticklabels(dates)
ax_top.tick_params(axis="x", labelsize=fontsize)
plt.setp(dates_top_axis, rotation = 335, ha = "right", rotation_mode = "anchor");
```

```python
lc = lc_gamma

# Plot flux PDF
fig_pdf, ax_pdf = lc.Plot_PDF()
ax_pdf.set_title("Flux PDF (%s)" % lc_name)
ax_pdf.set_xlabel("Photon Flux ($0.1$ GeV$<$E$<100$ GeV)\n[$10^{-6}$ cm$^{-2}$ s$^{-1}$]")
ax_pdf.set_ylabel("PDF")
# Plot PSD
fig_psd, ax_psd = lc.Plot_Periodogram()
ax_psd.set_title("Periodogram (%s)" % lc_name)
ax_psd.set_xlabel("Angular frequency (days$^{-1}$)")
ax_psd.set_ylabel("PSD")
```

## Fit of the PDF

The default PDF fit function in DELCgen is a linear combination of a st.gamma and a st.lognorm pdfs.
 - The n_args argument holds the list of the number of parameters of the functions within functions. 
 - Frozen is a list with the index(indices) of the parameter(s) to be fixed and corresponding value(s).

The shape of `st.gamma.pdf(x)` depends on 3 arguments: `a, loc=0, scale=1`.

The shape of `st.lognorm.pdf(x)` depends on 3 arguments: `s, loc=0, scale=1`, where s represents the standard deviation of the logarithms of the lightcurve fluxes.

The Fit_PSD method takes a vector af size given by `np.sum(n_args) = 6`, minus the number the parameters that were fixed, plus a weight factor for each PDF.

The `initial_params` in the `Fit_PDF` method has the following structure:
`[*free parameters of first function, *free parameters of second functions, weight first function, weight second function]`

Note that the post-fit weights $w_1$ and $w_2$ should be re-weighted to $\frac{w_1}{w_1 + w_2}$ and $\frac{w_2}{w_1 + w_2}$, respectively, when composing the total pdf.

Here, we try to use a single LogNorm function.

```python
fit_function = Mixture_Dist(
    functions = [st.lognorm],
    n_args = [3],
    frozen = False, # all parameters free
)

lc.Fit_PDF(
    initial_params = [0.9, -0.1, 0.3, 1.],
    model = fit_function
)

fig_pdf, ax_pdf = lc.Plot_PDF()
ax_pdf.set_xlabel("Photon Flux ($0.1$ GeV$<$E$<100$ GeV)\n[$10^{-6}$ cm$^{-2}$ s$^{-1}$]")
ax_pdf.set_ylabel("Frequency")
ax_pdf.grid(which = "both", color = "gray")
#ax_pdf.set_yscale("log")
#ax_pdf.set_ylim(bottom = 1e-4, top = 1)
plt.show()
```

For some unclear reasons, the fit implemented in DELCgen3 of the PDF using a single lognorm does not converge.
I will try to implementent different algorithms for the fit (e.g. iminuit), which hopefully converge more easily.

For now, for the scope of this notebook, I can use a set of parameters of the lognorm which describes approximately the data.

```python
fit_function = Mixture_Dist(
    functions = [st.lognorm],
    n_args = [3],
    frozen = False, # all parameters free
)
lc.pdfModel = fit_function
lc.pdfFit = {}
lc.pdfFit["x"] = [0.9, 0.0, 0.3, 1.0]
fig, ax = lc.Plot_PDF()
ax.set_yscale("log")
```

```python
values, bins, _ = plt.hist(lc.flux, bins = OptBins(lc.flux), density = True, label = "Data", alpha = 0.7)
xx = np.arange(0.00, 6, 0.01)
yyy = st.lognorm.pdf(
    x = xx,
    s = 0.9,
    loc = 0.0,
    scale = 0.3,
)
plt.plot(xx, yyy , label = "LogNorm", lw = 2)

plt.grid();
plt.yscale("log")
#plt.ylim(bottom = 1e-4, top = 1)
plt.legend();
```

### Best-fit PDF model for light-curve simulations

```python
PDFmodel = Mixture_Dist(
    functions = [st.lognorm],
    n_args = [3],
    frozen = False,
)

PDFparams = [0.9, 0.0, 0.3, 1.0]
```

## Fit of PSD

### First trial: power-law function

```python
fit_function = PL_no_offset
lc = lc_gamma

lc.Fit_PSD(initial_params = [2.5e-4, 1.5], model = fit_function)
fig_psd, ax_psd = lc.Plot_Periodogram()
ax_psd.set_title("Periodogram (%s)" % lc_name)
ax_psd.set_xlabel("Angular frequency (days$^{-1}$)")
ax_psd.set_ylabel("PSD")
text = "Power-law fit function\nPSD($\omega$) = A * $\omega$ $^{-\gamma}$\n"
text += "A = %.2e\n$\gamma$ = %.2f" % (lc.psdFit['x'][0], lc.psdFit['x'][1])
ax_psd.text(0.12, 0.2, text, transform=ax_psd.transAxes)
plt.show()

pl_no_offset_params = lc.psdFit["x"]
```

### Second trial: power-law function + constant

```python
fit_function = PL_with_offset
lc = lc_gamma

lc.Fit_PSD(initial_params = [1.e-5, 1.1, 20.e-4], model = fit_function)
fig_psd, ax_psd = lc.Plot_Periodogram()
ax_psd.set_title("Periodogram")
ax_psd.set_xlabel("Angular frequency (days$^{-1}$)")
ax_psd.set_ylabel("PSD")
text = "Power-law + offset fit function\nPSD($\omega$) = A * $\omega$ $^{-\gamma}$ + c\n"
text += "A = %.2e\n$\gamma$ = %.2f\nc = %.2e" % (lc.psdFit['x'][0], lc.psdFit['x'][1], lc.psdFit['x'][2])
ax_psd.text(0.12, 0.2, text, transform=ax_psd.transAxes);
```

### Best-fit PSD model for light-curve simulations

```python
### Best-fit model of the PSD LC
PSDmodel = PL_no_offset
PSDparams = pl_no_offset_params
```

## Light-curve simulations

```python
show_LC = True
save_LC = True

for i in range(1):
    print ("Generating LC number %d" % i)
    fake_LC = Simulate_DE_Lightcurve(
        PSDmodel = PSDmodel, 
        PSDparams = PSDparams,
        PDFmodel = PDFmodel,
        PDFparams = PDFparams,
        tbin = 3,
        LClength = int((mjd[-1] - mjd[0])/3),
    )
    
    daily_time, daily_flux = fake_LC.time.astype("float64") + np.min(mjd), fake_LC.flux
    flux_gamma = np.interp(mjd, daily_time, daily_flux)
    
    if save_LC:
        os.system("mkdir -p ./fake_LCs/Gamma/")
        np.savetxt("fake_LCs/Gamma/LC_gamma_%06d.txt" % i, np.transpose([mjd, flux_gamma]), delimiter = ",")
    
    if show_LC:
        fig, ax = plt.subplots(figsize = (15, 8))
        ax.plot(daily_time, daily_flux, ls = "None", marker = "o", ms = 2, label = "LC with 3 days bins")
        ax.set_title("Fake HE $\gamma$-ray lightcurve (E. method)")
        ax.plot(mjd, flux_gamma, ls = "None", marker = "o", ms = 4, label = "Same sampling as true LC")
        ax.set_xlabel("Time [MJD]")
        ax.set_ylabel("Photon Flux ($0.1$ GeV$<$E$<100$ GeV)\n[$10^{-6}$ cm$^{-2}$ s$^{-1}$]")
        ax.legend()
        ax.grid()
        plt.show()
```
